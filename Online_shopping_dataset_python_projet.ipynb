```python
# 1.je vais Analyser des produits les plus vendus en utilisant le product description et la quantity
top_produits = donnees_sans_nuls.groupby('Product_Description')['Quantity'].sum().nlargest(20)
print("Produits les plus vendus :\n", top_produits)
#le produit le plus vendu c'est le Maze Pen la quantité vendu c'est  16234.0 et le moins vendu des 20 c'est le Galaxy Screen Cleaning Cloth avec une quantité de  2801.0 
```

4.b.Visualiser les résultats de l’analyse puis fournir des observations pour expliquer les graphiques

```python
#voici un tableau qui va nous aider a mieux visualiser nos resultats
import matplotlib.pyplot as plt
plt.figure(figsize=(20, 6))
top_produits.plot(kind='bar', color='skyblue')
plt.title('Top 10 des Produits les Plus Vendus')
plt.xlabel('Produit')
plt.ylabel('Quantité Vendue')
plt.show()
```

L'analyse et le graphique montrent les top 20 des produits les plus vendus et le Maze pen est le produit le plus vendu  et la quantité vendu est de 16234 et le moins vendu des 20 c'est le Galaxy Screen Cleaning Cloth avec une quantite de  2801.0 

# 2.Pourquoi ? Peut-on conclure que le prix influence la quantité vendu ?

```python
# Je vais faire Création d'un DataFrame pour voire la relation entre  les prix et les quantités par produit
prix_produits = donnees_sans_nuls.groupby('Product_Description')['Avg_Price'].mean()
quantities_products = donnees_sans_nuls.groupby('Product_Description')['Quantity'].sum()
produit_data = pd.DataFrame({'Avg_Price': prix_produits, 'Quantity': quantities_products })
produit_data_top_20 = produit_data.nlargest(20, 'Quantity')
print(produit_data_top_20)
```

```python
#b. Visualiser les résultats de l’analyse puis fournir des observations pour expliquer les graphiques
#  voici une Visualisation en utilisant scatter plot  (Prix moyen vs Quantité vendue)
plt.scatter(produit_data_top_20['Avg_Price'], produit_data_top_20['Quantity'], alpha=0.5)
plt.title('Relation entre le prix moyen et la quantité vendue par produit')
plt.xlabel('Prix moyen')
plt.ylabel('Quantité vendue')
plt.grid(True)
plt.show()
```

L'analyse que nous venons de réaliser confirme qu'il y a une relation entre le prix du produit et la quantité car effecivement Maze Pen a un avg_price qui est bas de  0.914733 , mais elle montre également qu'il existe un autre facteur qui influence les ventes. Par exemple, le produit 'Nest Learning Thermostat 3rd Gen-USA' a un prix moyen de 150.981874, mais il se vend davantage que le produit 'Galaxy Screen Cleaning Cloth' qui a un prix moyen de 1.873281 et est moins vendu. Nous allons maintenant effectuer une analyse pour déterminer si la catégorie de produit influence également la quantité de produits vendus

```python
# Je vais faire l' analyse de la catégorie de produits la plus populaire
# Catégorie de produit par  categorie de produit
categorie_produits = donnees_sans_nuls.groupby('Product_Description')['Product_Category'].first()

# Création du DataFrame produit_data_2 avec les prix, les quantités et les catégories
produit_data_2 = pd.DataFrame({
    'Avg_Price': prix_produits,
    'Quantity': quantities_products,
    'Product_Category': categorie_produits
})

# Sélection des 20 produits les plus vendus par quantité
produit_data_2_top_20 = produit_data_2.nlargest(20, 'Quantity')

# Affichage du DataFrame avec la catégorie de produit ajoutée
print("Produits les plus vendus par quantité avec catégorie de produit :\n", produit_data_2_top_20)



```

```python
##b.Visualiser les résultats de l’analyse puis fournir des observations pour expliquer les graphiques
# Créer un graphique Pour visualiser la catégorie de produits la plus populaire
plt.figure(figsize=(10, 10))
plt.pie(produit_data_2_top_20['Quantity'], labels=produit_data_2_top_20['Product_Category'], autopct='%1.1f%%', startangle=140, colors=plt.cm.Paired.colors)
plt.title('Répartition des 20 Produits les Plus Vendus en Fonction du prix et de la Catégorie de produits')
plt.show()
```

L'analyse et le graphique confirment l'influence de la catégorie des produits sur la quantité d'articles vendus. Bien que les produits de la catégorie 'NEST-USA' soient relativement coûteux, ils connaissent une demande bien supérieure à celle de certains produits moins onéreux. Par exemple, le stylo à bille bleu, affiché à 0.899444 et classé dans la catégorie 'Office', enregistre des ventes inférieures à celles des produits 'NEST-USA', tels que le Nest Learning Thermostat 3rd Gen-USA - Stainless et le Nest Cam Indoor Security Camera - USA, vendus respectivement à 150.981874 et 120.214594. Ces constatations suggèrent que le prix et la catégorie de produit exercent tous deux une influence sur la quantité d'articles achetés. La catégorie 'Office', par exemple, représente la catégorie la plus achetée, bien que ses produits soient les moins chers, avec une part de marché de 13.5%. Néanmoins, des exceptions notables existent, comme les produits de la catégorie 'NEST-USA', qui, malgré leur coût élevé, surpassent en ventes certains produits de la catégorie 'Office'.

# 5.Conclure l’analyse en déduisant la présence/absence de lien de corrélation entre deux colonnes.

```python
#  je vais comparé Deux colonnes : quantité et prix 

colonnes_selectionnees = donnees_sans_nuls[['Quantity', 'Avg_Price']]

# Calcul de la corrélation entre les deux colonnes
correlation = colonnes_selectionnees.corr().iloc[0, 1]

# Interprétation de la corrélation
if correlation > 0:
    interpretation = "la corrélation est  positive."
elif correlation < 0:
    interpretation = "la corrélation est  négative."
else:
    interpretation = "Il n'y a pas de corrélation."
    
```

```python
print(f"la Corrélation entre 'Quantity' et 'Avg_Price' : {correlation}")
print(interpretation)
```

La valeur de la corrélation entre 'Quantity' et 'Avg_Price' est -0.1141, indiquant une corrélation négative entre la quantité achetée et le prix moyen. Cela suggère que que  lorsque le prix moyen augmente, la quantité achetée a tendance à diminuer, et vice versa. Une corrélation négative indique une relation inverse entre les deux variables.
Donc on peut conclure notre analyse en disant  que la catégorie de produits et le prix moyen jouent tous deux un rôle crucial dans les habitudes d'achat. Bien que des catégories moins chères comme 'Office' soient généralement populaires, des exceptions comme 'NEST-USA' démontrent que d'autres facteurs, tels que la categorie perçue, peuvent influencer les choix des consommateurs. La corrélation négative entre la quantité et le prix renforce cette idée en suggérant une tendance globale où une hausse des prix est associée à une diminution des quantités achetées.
